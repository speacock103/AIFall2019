```
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import random

#this simple funciton contains a minimum, with graddient zero
def function1(x,y): 
    return x**2 + y**2

#here the 3d contour is plotted
x1 = np.linspace(-10, 10, 20)
y1 = np.linspace(-10, 10, 20)
X1, Y1 = np.meshgrid(x1, y1)
Z1 = function1(X1, Y1)

fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X1, Y1, Z1, 50)
```

```
#next the 2d contour plot of function 1 is plotted
fig2 = plt.figure()
ax = fig2.add_subplot(111)

ax.contour(X1,Y1,Z1)

plt.show()
```

```
#next we will define a simple function that has a maximum with gradient zero
def function2(x,y): 
    return - x**2 - y**2

#here the 3d contour is plotted
x2 = np.linspace(-10, 10, 20)
y2 = np.linspace(-10, 10, 20)
X2, Y2 = np.meshgrid(x2, y2)
Z2 = function2(X2, Y2)

fig3 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X2, Y2, Z2, 50)
```

```
#next the 2d contour plot of function 2 is plotted
fig4 = plt.figure()
ax = fig4.add_subplot(111)

ax.contour(X2,Y2,Z2)

plt.show()
```

```
#next we will define a simple function with a gradient of zero at a point that
#is neither a max or min
def function3(x,y): 
    return x**2 - y**2

#here the 3d contour is plotted
x3 = np.linspace(-10, 10, 20)
y3 = np.linspace(-10, 10, 20)
X3, Y3 = np.meshgrid(x3, y3)
Z3 = function3(X3, Y3)

fig5 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X3, Y3, Z3, 50)
```

```
#lastly the 2d contour plot of function 3 is plotted
fig6 = plt.figure()
ax = fig6.add_subplot(111)

ax.contour(X3,Y3,Z3)

plt.show()
```
