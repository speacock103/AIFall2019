```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import random
```

```
import numpy as np
np.random.seed(42)

def function(x1, x2):
    return 3 + 4*x1 - x2 + np.random.rand(100, 1)

xs1 = 2 * np.random.rand(100, 1)
xs2 = 4 * np.random.rand(100, 1)
ys = function(xs1, xs2)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(xs1, xs2, ys, c='b');
```

```
# train set
train_xs1 = xs1[:80]
train_xs2 = xs2[:80]
train_ys = ys[:80]
# test set
test_xs1 = xs1[80:]
test_xs2 = xs2[80:]
test_ys = ys[80:]
```

```
# number of epochs
epochs = 10
# learning rate
lr = 0.0003
```

```
# initial value for weight w and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_xs1[i] + w2 *train_xs2[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_xs1[i]
    grad_w2 = (y_pred - train_ys[i]) * train_xs2[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += (w1 * test_xs1[i] + w2 * test_xs2[i] + b - test_ys[i]) ** 2
test_loss /= 20
```

```
pred_ys = w1 * test_xs1 + w2 * test_xs2 + b
```

```
fig2 = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(test_xs1, test_xs2, test_ys, c='b');
ax.scatter3D(test_xs1, test_xs2, pred_ys, c='r');
```
